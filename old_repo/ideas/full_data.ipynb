```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read full data
full_data = pd.read_csv('../data/eq2015/building_structure/csv_building_structure.csv')
full_data.shape
```

```python
# Exploring columns related to the building characteristics
cols_building_type = [col for col in full_data.columns if '_type' in col]
building_type = full_data[cols_building_type].copy()
building_type.head()
```

```python
for col in building_type.columns:
    print(building_type[col].value_counts(), end='\n\n')
```

```python
from matplotlib.pyplot import tight_layout

fig, axs = plt.subplots(2, 2, figsize=(10, 6), tight_layout=True)
for i, col in enumerate(building_type.columns):
    sns.barplot(x=building_type[col].value_counts(),
                y=building_type[col].value_counts().index,
                ax=axs.flatten()[i])
```
