**GOAL**: predict the level of damage a building 'building_id' suffered as a result of the 2015 earthquake  
- 1 represents low damage
- 2 represents a medium amount of damage
- 3 represents almost complete destruction

```python
# Load data
train_data = load_train_data(local=True)
train_data.shape
```

```python
# Experiment with BaseN encoder
categorical_values = train_data.select_dtypes(include='object')
columns_categorical = categorical_values.columns

from category_encoders import BaseNEncoder

basen_enc = BaseNEncoder(drop_invariant=True, base=1)

X_cat = categorical_values.copy()

basen_enc.fit_transform(X_cat)
```

```python
from sklearn.preprocessing import TargetEncoder

target_encoder = TargetEncoder(target_type='multiclass')

col_to_encode = ['geo_level_1_id']

X_enc = train_data[col_to_encode].copy()
y_enc = train_data['damage_grade'].copy()

target_encoder.fit(X_enc, y_enc)
target_encoder.transform(X_enc)
```
